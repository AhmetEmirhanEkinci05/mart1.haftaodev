```python
import pandas as pd
1
# Örnek CSV dosyası oluşturmak 
data={"isim": ["Emirhan","Bora","Aslı","Deniz","Arda"],
      "yas": [19,20,33,22,28], 
      "sehir": ["istanbul","Ankara","Bursa","Eskişehir","İzmir"]}
df = pd.DataFrame(data)
df.to_csv("ornek.csv", index=False)

# CSV dosyasını okuma 
df = pd.read_csv("ornek.csv")
print(df)

2
# İlk 2 satırı gösterme
print(df.head(2))
# Son 2 satırı gösterme
print(df.tail(2))
3
# veri çerçevesi hakkında bilgi
print(df.info())
4
#sayısal verilerin istatiksel özeti
print(df.describe())
5
#"isim" sütununu seç
print(df["isim"])
# ilk satırı seç
print(df.loc[0])
6
# örnek bir veri çerçevisinde eksik değer
df.loc[1,"yas"] = None
print(df)
# Eksik değerleri 0 ile doldur
df_dolu = df.fillna(0)
print(df_dolu)
7
# Şehrirlere göre yaş ortalamasını bulma
yas_ortalaması = df.groupby("sehir")["yas"].mean()
print(yas_ortalaması)
8
# Şehir sütununu silme
df_yeni = df.drop("sehir", axis=1)
print(df_yeni)
9
# "Şehir" sütunundaki benzersiz değerler
benzersiz_sehirler = df["sehir"].unique()
print(f"Benzersiz Şehirler:{benzersiz_sehirler}")

10 
# Yaşa göre sıralama
df_sirali = df.sort_values(by="yas")
print(df_sirali)
# ilk satırın "yas" değerini seç
print(df.loc[0,"yas"])
# ilk 2 satırı ve "isim" ve "yas" sütunla seç
print(df.loc[0:1,["isim","yas"]])
# ilk 2 satırı ve ilk 2 sütunu seç
print(df.iloc[0:2,0:2])

       
```
